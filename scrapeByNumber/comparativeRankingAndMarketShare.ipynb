#### 2.Calculate today's ranking interms of previous day's ranking

```python
# Find the change in today's rank  compared to yesterday's rank
# Function to trace ranking
def traceRanking(yes,today):
    #Read today's and yesterday's data
    yesDf = pd.read_csv(f'{yes}_top_500_courses.csv')
    todayDf = pd.read_csv(f'{today}_top_500_courses.csv')
    
    # Insert ranking
    yesDf.insert(loc=1,value=np.arange(1,501),column='ranking')
    todayDf.insert(loc=1,value=np.arange(1,501),column='ranking')
    
    #Drop duplicates by ids
    yesDf.drop_duplicates(subset ='courseId',keep=False,inplace=True)
    todayDf.drop_duplicates(subset ='courseId',keep=False,inplace=True)
    
    # Set these 3 columns as index
    yesDf = yesDf.set_index(['courseId','courseTitle','courseProvider'])
    todayDf = todayDf.set_index(['courseId','courseTitle','courseProvider'])
    
    # Match yesterday's index to today's
    setYesToToday = yesDf.reindex_like(todayDf)
    
    # Change of today interms of yesterday. + indicates advancement from yesterday, - indicates drop from yesterday.
    rankChanged = setYesToToday.ranking - todayDf.ranking
    # Make column from index
    rankChanged = rankChanged.reset_index(level=['courseTitle','courseProvider'])
    #Change column name
    rankChanged = rankChanged.rename({'ranking':'rankChanged'},axis=1)
    # Fill nan with new
    rankChanged.rankChanged.fillna('new',inplace=True)
    #Create the csv file
    return rankChanged.to_csv(f'{today}_rank_change.csv')  
```
