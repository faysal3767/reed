#### In this project we will try to find out different certification effects on course revenue between two dataframes of different dates on Reed

```python
# Import basic modules
import pandas as pd
import numpy as np
```

```python
# Read to two csv files to to tack unit sale and actual revenue
preDf = pd.read_csv("11_Mar_15_providers.csv")
nextDf = pd.read_csv("19_Mar_15_providers.csv")
nextDf.head()
```

```python
# Merge two dataframes on course ids
mergedDf = pd.merge(nextDf,preDf,on="courseId",how="left",indicator=True)
# Common courses on both data
mergedBothDf = mergedDf[mergedDf._merge=="both"]
mergedBothDf.columns
```

```python
# Column length to insert no of unit sold of course at the at of the dataframe
columnLengthForUnitSold = len(mergedBothDf.columns.values)

# Create a column name "unitSold" between two time duration at the the end of the mergedReqDf
mergedBothDf.insert(loc=columnLengthForUnitSold,value=mergedBothDf.unitSold_x-mergedBothDf.unitSold_y, column="unitSold")

# List to store actual revenue.
# According to Reed, the revenue earned by a provider varies according to following price s
actualRevenue = []
totalMissingOriginalPriceNextDay = mergedBothDf[mergedBothDf.originalPrice_x.isna()].courseProvider_x.value_counts().sum()
totalMissingOriginalPricePreDay = mergedBothDf[mergedBothDf.originalPrice_y.isna()].courseProvider_y.value_counts().sum()

if totalMissingOriginalPriceNextDay<totalMissingOriginalPricePreDay:
    for prc,sale in zip(mergedBothDf.offerPrice_x,mergedBothDf.unitSold):
        if 10<=prc<100:
            actualRevenue.append(prc*0.4*sale)
        elif 100<=prc<150:
            actualRevenue.append(prc*0.55*sale)

        elif 150<=prc<200:
            actualRevenue.append(prc*0.6*sale)

        elif 200<=prc<300:
            actualRevenue.append(prc*0.7*sale)

        elif 300<=prc<500:
            actualRevenue.append(prc*0.75*sale)

        elif 500<=prc:
            actualRevenue.append(prc*0.8*sale)
else:
    for prc,sale in zip(mergedBothDf.offerPrice_y,mergedBothDf.unitSold):
        if 10<=prc<100:
            actualRevenue.append(prc*0.4*sale)
        elif 100<=prc<150:
            actualRevenue.append(prc*0.55*sale)

        elif 150<=prc<200:
            actualRevenue.append(prc*0.6*sale)

        elif 200<=prc<300:
            actualRevenue.append(prc*0.7*sale)

        elif 300<=prc<500:
            actualRevenue.append(prc*0.75*sale)

        elif 500<=prc:
            actualRevenue.append(prc*0.8*sale)

# Create a new colum named actual revenue at the end of the current df
columnLengthForActualRevenue = len(mergedBothDf.columns.values)
mergedBothDf.insert(loc=columnLengthForActualRevenue,value=actualRevenue,column="actualRevenue")
```

```python
# Insert another column named "totalQualification" combining cpdAccreditedBy and awrbodyname
totalQual = mergedBothDf.cpdAccreditedBy_y.str.strip() + "+" + mergedBothDf.awrBodyName_x
mergedBothDf.insert(loc=5,value=totalQual,column="totalQual")
```

```python
# Lets look at those 3 columns
mergedBothDf[["totalQual","cpdAccreditedBy_y","awrBodyName_x"]].head()
```

```python
# Frequency of total qualifications
mergedBothDf["totalQual"].value_counts()
```

```python
# Clean total qualification column
# Remove +missing and fillna with "missing". Note that backward slash
mergedBothDf.loc[:,"totalQual"] = mergedBothDf.loc[:,"totalQual"].str.replace("\+missing","").fillna("missing")
mergedBothDf.totalQual.value_counts()
```

```python
# Remove duplicates using set
mergedBothDf.totalQual = mergedBothDf.totalQual.str.split("+").map(set).astype(str).str.replace("{","").astype(str).str.replace("}","")
```

```python
# This is the cleaned totalQual
mergedBothDf.totalQual.value_counts()
```

```python
# Checkout the column again
mergedBothDf.head()
```

```python
# Extract the required columns
reqDf = mergedBothDf.loc[:,["courseId",
                            "courseTitle_x",
                            "subTitle_x",
                            "courseLink_x",
                            "courseProvider_x",
                            "totalQual",
                            "offerPrice_x",
                            "unitSold",
                            "actualRevenue"
                            ]]
# Remove "_x" from column names
reqDf.columns = reqDf.columns.str.replace("_x","")

# And checkout the final df
reqDf.head()
```

```python
# No of courses by awarding bodies
reqDf.groupby("totalQual").courseTitle.count().sort_values(ascending=False).to_frame("totalCourses").head(10)
```

```python
# Total revenue across different certifications and providers. Make totalRevenue and courseProvider column again
groupedRevenueByTotalQualAndProvider = reqDf.groupby(["totalQual","courseProvider"]).actualRevenue.sum().sort_values(ascending=False).to_frame().reset_index(level=["courseProvider","totalQual"]).round()
groupedRevenueByTotalQualAndProvider.head()
```

```python
# Create a pivot table with values actualRevenue and make courseProvider column again
pivotGroupedRevenueByTotalQualAndProvider = groupedRevenueByTotalQualAndProvider.pivot_table(index="courseProvider", columns="totalQual", values="actualRevenue", fill_value=0).reset_index(level="courseProvider")
pivotGroupedRevenueByTotalQualAndProvider
```

```python
# Create a csv file 
pivotGroupedRevenueByTotalQualAndProvider.to_csv("revenueByCertification.csv",index=None)
```
